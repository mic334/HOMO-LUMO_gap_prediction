```python
from rdkit import Chem


suppl = Chem.SDMolSupplier("../data/ZINC_casual_raw/Compound_000000001_000500000.sdf")

smiles = []
for mol in suppl:
    if mol is None:
        continue
    smi = Chem.MolToSmiles(mol)
    smiles.append(smi)
    
    if len(smiles) == 10000:
        break

print(smiles[:5])
```
